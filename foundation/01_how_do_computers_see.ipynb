# Importing Images In Python

This notebook is a primer for importing images in Python and using visualizing how an image fundamentally looks as a set of matrices of numbers.

```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading a sample image
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//images.jpg')
```

```python
# Displaying the original image
plt.imshow(img)
plt.title('Image of the Indian ruppee sign')
plt.show()
```

```python
# Converting image to grascale and resizing it
gr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # converting the image into grayscale
gr = gr[:215, :] # trimming the edges
gr = imutils.resize(gr, 20) # resizing the image to fit in screen
gr = 255 - gr # inverting the image
gr[gr > 10 ] = 255 # applying threshold
gr[gr <= 10] = 0 # applying threshold
```

```python
# Displaying the grascale image
plt.imshow(gr, cmap='gray', vmin=0, vmax=255)
plt.title('Image of the Indian ruppee sign resized to 20 X 20')
plt.show()
```

```python
# Display the RGB channels of the image seperately
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//rgb_colors.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Colorful image')
plt.show()
```

```python
# Displaying RGB channels individually as grayscale images
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (50,10))
# adjust spacing and file size
f.suptitle('RGB Channels visualized seperately (a) Red (b) Green (c) Blue', fontsize=60)
ax1.imshow(img[:,:,0], cmap='gray', vmin=0, vmax=255)
ax1.set_title('Red Channel',fontsize=32)
ax2.imshow(img[:,:,1], cmap='gray', vmin=0, vmax=255)
ax2.set_title('Green Channel',fontsize=32)
ax3.imshow(img[:,:,2], cmap='gray', vmin=0, vmax=255)
ax3.set_title('Blue Channel',fontsize=32)
plt.show()
```
