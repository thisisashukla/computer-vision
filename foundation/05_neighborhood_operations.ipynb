```python
# Importing libraries
import cv2
import sys
import imutils
sys.path.append('../')
from basics import functions as fn
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
img = cv2.imread("C://Users//ankushukla//Documents//Work//CV Training//taj.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
average3 = np.ones((3,3))/9
average5 = np.ones((5,5))/25

averImg3 = cv2.filter2D(img, -1, average3)
averImg5 = cv2.filter2D(img, -1, average5)

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (50,50))
ax1.imshow(img, cmap='gray', vmin=0, vmax=255)
ax2.imshow(averImg3, cmap='gray', vmin=0, vmax=255)
ax3.imshow(averImg5, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
gaussianKernel = np.array([[1,2,1], [2,4,2],[1,2,1]])/16

gaussianImage = cv2.filter2D(img, -1, gaussianKernel)
gaussianBlur = cv2.GaussianBlur(img, (7,7), 0)
medianImg = cv2.medianBlur(img, 7)

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (50,50))
ax1.imshow(gaussianImage, cmap='gray', vmin=0, vmax=255)
ax2.imshow(gaussianBlur, cmap='gray', vmin=0, vmax=255)
ax3.imshow(medianImg, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
sobelx = np.array([[1,2,1], [0,0,0],[-1,-2,-1]])
sobely = np.array([[-1,0,1], [-2,0,2],[-1,0,-1]])
laplacian = np.array([[0,-1,0], [-1,4,-1],[0,-1,0]])
print(sobelx)
print(sobely)
print(laplacian)
gradientx = cv2.filter2D(img, -1, sobelx)
gradienty = cv2.filter2D(img, -1, sobely)
lapImg = cv2.filter2D(img, -1, laplacian)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize = (50,50))
ax1.imshow(img, cmap='gray', vmin=0, vmax=255)
ax2.imshow(gradientx, cmap='gray', vmin=0, vmax=255)
ax3.imshow(gradienty, cmap='gray', vmin=0, vmax=255)
ax4.imshow(lapImg, cmap='gray', vmin=0, vmax=255)
plt.show()
```
